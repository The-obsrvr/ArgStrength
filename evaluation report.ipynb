```python
import numpy as np
import pandas as pd
```

### Evaluation for Task 1

```python
gretz_eval = pd.read_csv("task1/gretz/eval_results.csv", sep=",")
swanson_eval = pd.read_csv("task1/swanson/eval_results.csv", sep=",")
ukprank_eval = pd.read_csv("task1/ukprank/eval_results.csv", sep=",")
toledo_eval = pd.read_csv("task1/toledo/eval_results.csv", sep=",")
```

```python
toledo_eval_agg = toledo_eval.groupby(["test_file"], as_index=False).agg(
                      {'pearson':['mean','std'], 'spearman':['mean','std']})
gretz_eval_agg = gretz_eval.groupby(["test_file"], as_index=False).agg(
                      {'pearson':['mean','std'], 'spearman':['mean','std']})
swanson_eval_agg = swanson_eval.groupby(["test_file"], as_index=False).agg(
                      {'pearson':['mean','std'], 'spearman':['mean','std']})
ukprank_eval_agg = ukprank_eval.groupby(["test_file"], as_index=False).agg(
                      {'pearson':['mean','std'], 'spearman':['mean','std']})
```
